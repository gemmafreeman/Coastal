```python
import pandas as pd
import matplotlib.pyplot as plt

# Define beaches and contours
beaches = ['Narrabeen-Mid', 'Narrabeen-South', 'Narrabeen-North', 'Maroubra', 'Tamarama']  # Replace with actual beach names
contours = ['10', '30']

# Create nested dictionary for data storage
data = {beach: {contour: None for contour in contours} for beach in beaches}

# Load all files systematically
for beach in beaches:
    for contour in contours:
        filename = f"WaveTimeSeries_{beach}_30May-27Jun2019_{contour}-metre-contour.csv"     
        df = pd.read_csv(filename, parse_dates=['UTC'], index_col='UTC')
        data[beach][contour] = df['2019-06-04':'2019-06-07']  # Storm filter
```
