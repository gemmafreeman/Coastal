```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
wave_data = pd.read_excel("2019_wavedata.xlsx", parse_dates=["Timestamp"])

# Check column names (adjust if needed)
print("Columns:", wave_data.columns.tolist())

# Plot Hsig and Hmax over time
plt.figure(figsize=(12, 6))
plt.plot(wave_data["Timestamp"], wave_data["Hsig"], label="Hsig (m)", alpha=0.8)
plt.plot(wave_data["Timestamp"], wave_data["Hmax"], label="Hmax (m)", alpha=0.6)
plt.title("2019 Manly Buoy Wave Height Time Series")
plt.xlabel("Date")
plt.ylabel("Wave Height (m)")
plt.legend()
plt.grid(True)
plt.show()

```
